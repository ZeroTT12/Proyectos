```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("colorado_final.csv")
# print(df.head())


features = [
    "sqft",
    "beds",
    "baths",
    "stories",
    "garage",
    "baths_full",
    "baths_full_calc"
]

df_knn = df[features]
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_knn)

imputer = KNNImputer(
    n_neighbors=4,      # número de vecinos
    weights="distance"  # más peso a vecinos más cercanos
)

df_imputed_scaled = imputer.fit_transform(df_scaled)


df_imputed = scaler.inverse_transform(df_imputed_scaled)
df_imputed = pd.DataFrame(df_imputed, columns=features)
# print(df_imputed.isnull().sum())

df[features] = df_imputed

```

```python
cols_int = ["sqft","beds","baths","stories","garage","baths_full","baths_full_calc"]
df[cols_int] = df[cols_int].round().astype(int)

df.to_csv("colorado_analisis.csv", index=False)
```
